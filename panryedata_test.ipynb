```python
import json
import csv
```

```python
f = open('C:/Users/jayts/OneDrive/바탕 화면/Tictag/json/KAS/panryedata.json', 'r', encoding='UTF8')
```

```python
info = json.load(f)
baselist = []
```

```python
info['cases'][0]['base']
```

```python
for i in range(0, len(info['cases'])):
    baselist.append(info['cases'][i]['base'])
```

```python
with open('C:/Users/jayts/OneDrive/바탕 화면/Tictag/json/KAS/base.csv', 'w', newline='') as file:
    fieldnames=['name', 'uen', 'type', 'date']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    for i in range(0, len(baselist)):
        writer.writerow(baselist[i])
```
