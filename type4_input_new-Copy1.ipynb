```python
import json
import csv
from tqdm.notebook import tnrange
import pandas as pd
import re
```

```python
input_repo = '/Users/jaytsol/Desktop/Type23 (2)/누적/accum_concat'
output_repo = '/Users/jaytsol/Desktop/Type4/test'

input_filename = '통합_검수본_21113_1120.csv'
output_filename = 'type4_input.csv'
```

```python
org_data = pd.read_csv(f'{input_repo}/{input_filename}')
```

```python
# type4_data가 
type4_input = pd.DataFrame(columns=['lineNumber', 'target', 'len', 'input_1(type4)', 'input_1(type5)', 'input_2(type4)', 'input_2(type5)', 'input_3(type4)', 'input_3(type5)'])
for i in tnrange(len(org_data)):
    transcription_without_asterisk = re.sub('(\<=.*?\>{1})','',str(org_data['transcription'][i])).replace('<<<', '').replace('>>>', '')
    if len(transcription_without_asterisk) < 200:
        pass
    else:
        type4_input = type4_input.append({'lineNumber':org_data['lineNumber'][i], 'target':transcription_without_asterisk, 'len':len(transcription_without_asterisk)}, ignore_index=True)
```
