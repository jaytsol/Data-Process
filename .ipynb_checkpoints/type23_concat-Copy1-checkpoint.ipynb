```python
import pandas as pd
from os import walk
import os
import xlrd
import csv
from tqdm.notebook import tnrange
```

```python
input_repo = '/Users/jaytsol/Desktop/Type23 (2)/임시/type D 작업물 정리'
output_repo = '/Users/jaytsol/Desktop/Type23 (2)/임시/데이터합쳐'

xlsx_output_repo = '/Users/jaytsol/Desktop/Type23 (2)/임시/D_output'
csv_input_repo = output_repo
```

```python
# 해당 경로의 파일리스트 추출
def get_filelist(input_repo):
    file_list = []
    for (dirpath, dirnames, filenames) in walk(input_repo):
        file_list.extend(filenames)
        break

    return file_list
```

```python
# 해당 경로의 csv파일의 경로+파일명 리스트 추출
def get_csvlist(input_repo, file_list):
    csv_list = []
    for i in range(len(file_list)):
        filename, file_extension = os.path.splitext(f'{input_repo}/{file_list[i]}')

        if file_extension == '.csv':
            csv_list.append(filename+file_extension)

    return csv_list
```

```python
# 해당 경로의 csv파일의 경로+파일명 리스트 추출
def get_xlsxlist(input_repo, file_list):
    xlsx_list = []
    for i in range(len(file_list)):
        filename, file_extension = os.path.splitext(f'{input_repo}/{file_list[i]}')

        if file_extension == '.xlsx':
            xlsx_list.append(filename+file_extension)

    return xlsx_list
```

```python
def concat_csvfile_1(csv_list):
    COLUMNS = ['input', 'type4', 'type5']
    result = pd.concat([pd.read_csv(f) for f in csv_list])    # 라인넘버 기준 안합친파일 합치기
    
    return result
```

```python
def concat_csvfile(csv_list):
    COLUMNS = ['input', 'type4', 'type5']
    result = pd.concat([pd.read_csv(f) for f in csv_list])    # 라인넘버 기준 안합친파일 합치기
    
    return result
```
