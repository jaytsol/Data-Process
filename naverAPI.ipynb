```python
import json
import pandas as pd
from tqdm import tnrange, tqdm_notebook
from hanspell import spell_checker
```

```python
# 문장 맞춤법 고쳐서 수정된 문장 출력.
def correction(ori_sent):
    spelled_sent = spell_checker.check(ori_sent)
    Error_Code = spelled_sent.words
    tmp_list = []
    for i, j in Error_Code.items():
        tmp_list.append(j)
        
    PASSED = tmp_list.count(0)
    WRONG_SPELLING = tmp_list.count(1)
    WRONG_SPACING = tmp_list.count(2)
    AMBIGUOUS = tmp_list.count(3)
    STATISTICAL_CORRECTION = tmp_list.count(4)
    Num_of_Error_Code = [PASSED, WRONG_SPELLING, WRONG_SPACING, AMBIGUOUS, STATISTICAL_CORRECTION]
    
    return spelled_sent.checked, spelled_sent.errors, Num_of_Error_Code
```

```python
result = org
result.insert(3, 'naverAPI', None)
result.insert(4, 'naverAPI_check', None)
result.insert(5, 'spelled_sent', None)
result.insert(6, 'error_code', None)
```

```python
for i in tnrange(len(result)):
    try:
        result['naverAPI'][i], result['spelled_sent'][i], result['error_code'][i] = correction(result['trans'][i])
        if result['naverAPI'][i] != result['trans'][i]:
            result['naverAPI_check'][i] = 'o'
    except:
        result['naverAPI'][i] = result['trans'][i][0]
        result['naverAPI_check'][i] = 'o'
```

```python
error_without_space_error = pd.DataFrame(columns=['resultNumber', 'word', 'usecase', 'trans', 'naverAPI', 'naverAPI_check', 'spelled_sent', 'error_code', 'keyword'])
for i in tnrange(len(result)):
    try:
        WRONG_SPELLING = result['error_code'][i][1]
        AMBIGUOUS = result['error_code'][i][3]
        STATISTICAL_CORRECTION = result['error_code'][i][4]
        if (WRONG_SPELLING != 0)|(AMBIGUOUS != 0)|(STATISTICAL_CORRECTION):
            error_without_space_error = error_without_space_error.append(result.loc[i], ignore_index=True)
    except:
        error_without_space_error = error_without_space_error.append(result.loc[i], ignore_index=True)
```
