```python
result = {"translate":[]}
for i in range(0, 75000):
    keyword = []
    origin = org['originalText'][i]
    trans = org['tag_removed'][i]
    keyword.append(org['word'][i])
    result['translate'].append({"origin":origin, "trans":trans, "keyword":keyword})
```
